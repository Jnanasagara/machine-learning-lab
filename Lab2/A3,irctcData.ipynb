```python
import pandas as pd
import statistics
import matplotlib.pyplot as plt

# Load and clean df
df = pd.read_csv('E:/COLLEGE PDFS/5TH SEMESTER/Machine Learning/irctcData.csv')
df = df[['Date', 'Month', 'Day', 'Price', 'Chg%']]  # Only keep necessary columns
df.dropna(inplace=True)  # Remove empty rows

# Convert 'Price' and 'Chg%' from strings to numerical formats
df['Price'] = df['Price'].str.replace(',', '').astype(float)
df['Chg%'] = df['Chg%'].str.replace('%', '').astype(float)

#Overall Mean and Variance of Price
all_prices = df['Price'].tolist()
overall_mean = statistics.mean(all_prices)
overall_variance = statistics.variance(all_prices)

#Wednesday Mean
wednesday_prices = df[df['Day'] == 'Wed']['Price'].tolist()
wednesday_mean = statistics.mean(wednesday_prices)

#April Mean
april_prices = df[df['Month'] == 'Apr']['Price'].tolist()
april_mean = statistics.mean(april_prices)

#Probability of Loss
loss_days = df[df['Chg%'] < 0]
prob_loss = len(loss_days) / len(df)

#Probability of Profit on Wednesday
wednesday_days = df[df['Day'] == 'Wed']
wednesday_profit_days = wednesday_days[wednesday_days['Chg%'] > 0]
prob_profit_wednesday = len(wednesday_profit_days) / len(wednesday_days)

#Conditional Probability of Profit Given Wednesday -- already done, since we are already filtering by Wednesdays

#scatter plot of change % by day of week ---
plt.figure(figsize=(10, 5))
plt.scatter(df['Day'], df['Chg%'], color='blue', alpha=0.6)
plt.title("Daily % Change vs Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Chg%")
plt.grid(True)
plt.tight_layout()
plt.show()

# Print outputs
print(f"Overall Mean Price: {overall_mean:.2f}")
print(f"Overall Variance in Price: {overall_variance:.2f}")
print(f"Average Price on Wednesdays: {wednesday_mean:.2f}")
print(f"Average Price in April: {april_mean:.2f}")
print(f"Probability of Making a Loss: {prob_loss:.3f}")
print(f"Probability of Profit on Wednesday: {prob_profit_wednesday:.3f}")
```
